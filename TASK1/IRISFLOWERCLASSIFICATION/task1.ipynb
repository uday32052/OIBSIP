```python
#ploting graph usning seaborn
sns.set_style('whitegrid')
sns.pairplot(data = df, hue='species')
```

```python

pet_len_wid = df[df.species == 'Iris-setosa'].plot(kind = 'scatter', x = 'sepal_length', y = 'sepal_width'
,color = 'blue', label = 'setosa')
df[df.species == 'Iris-versicolor'].plot(kind = 'scatter', x = 'sepal_length', y = 'sepal_width', color = 'red'
,label = 'versicolor', ax = pet_len_wid)
df[df.species == 'Iris-virginica'].plot(kind = 'scatter', x = 'sepal_length', y = 'sepal_width', color = 'green'
,label = 'virginica', ax = pet_len_wid)
pet_len_wid.set_xlabel('sepal_length')
pet_len_wid.set_ylabel('sepal_width')
pet_len_wid.set_title('sepal-length-width Vs species')
pet_len_wid = plt.gcf()
pet_len_wid.set_size_inches(10, 5)
plt.show()
```

```python
pet_len_wid = df[df.species == 'Iris-setosa'].plot(kind = 'scatter', x = 'petal_length', y = 'petal_width'
,color = 'blue', label = 'setosa')
df[df.species == 'Iris-versicolor'].plot(kind = 'scatter', x = 'petal_length', y = 'petal_width', color = 'red'
,label = 'versicolor', ax = pet_len_wid)
df[df.species == 'Iris-virginica'].plot(kind = 'scatter', x = 'petal_length', y = 'petal_width', color = 'green'
,label = 'virginica', ax = pet_len_wid)
pet_len_wid.set_xlabel('petal_length')
pet_len_wid.set_ylabel('petal_width')
pet_len_wid.set_title('petal-length-width Vs species')
pet_len_wid = plt.gcf()
pet_len_wid.set_size_inches(10, 5)
plt.show()
```
